```python
import requests
import pandas as pd
import json
```

```python
imdb_api_key = 'k_nf72o54n'
movie_title = ['Pitch Perfect', 'Frozen', 'The Notebook', 'Parasite', 'The Godfather', 'The Ring', 'Scary Movie', 'The Avengers',
              'Titanic', 'Brave']
stars_list = []
ratings_list = []
directors_list = []
genres_list = []
imdb_rating_list = []
release_date_list = []
runtime_list = []

for movies in movie_title:
    imdb_url = f'https://imdb-api.com/en/API/SearchTitle/{imdb_api_key}/{movies}'
    imdb_req = requests.get(imdb_url)
    imdb_data = imdb_req.json()
    
    imdb_id = imdb_data["results"][0]["id"]

    imdb_detail_url = f'https://imdb-api.com/en/API/Title/{imdb_api_key}/{imdb_id}'
    detail_req = requests.get(imdb_detail_url)
    detail_data = detail_req.json()
    
    stars_list.append(detail_data["stars"])
    ratings_list.append(detail_data["contentRating"])
    directors_list.append(detail_data["directors"])
    genres_list.append(detail_data["genres"])
    imdb_rating_list.append(float(detail_data['imDbRating']))
    release_date_list.append(detail_data['releaseDate'])
    runtime_list.append(int(detail_data['runtimeMins']))

    print(f'Movie Title: {movies}\nGenre: {genres_list}\nContent Rating: {ratings_list}\nStars: {stars_list}\nDirector: {directors_list}')
    print(f'Release Date: {release_date_list}\nRuntime: {runtime_list}\nIMDb Rating: {imdb_rating_list}\n')
```

```python
movies_df = pd.DataFrame()

movies_df['Title'] = movie_title
movies_df['Stars'] = stars_list
movies_df['Ratings'] = ratings_list
movies_df['Directors'] = directors_list
movies_df['Genre'] = genres_list
movies_df['IMDb Rating'] = imdb_rating_list
movies_df['Release Date'] = release_date_list
movies_df['Runtime'] = runtime_list

movies_df.to_csv('IMDb Ratings.csv', index = False)
movies_df
```

```python
imdb_ratings_stats = movies_df['IMDb Rating'].describe()
runtime_stats = movies_df['Runtime'].describe()

stats_dictionary = {'IMDb Rating': imdb_ratings_stats, 'Runtime': runtime_stats}
stats_df = pd.DataFrame(stats_dictionary)
stats_df
```
